Importing the dependencies


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn.metrics import accuracy_score
```

```python
from io import StringIO # Use io module instead of StringIO
```

Importing dataset


```python
dataset = pd.read_csv('housing.csv', sep='\s+', header=None)
```

```python
print(dataset)
```

```python
dataset.shape
```

```python
#Loading the dataset to Pandas Dataframe
```

```python
col_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'Price']
```

```python
# Convert the DataFrame to a CSV string
dataset_str = dataset.to_csv(index=False, header=None, sep=" ") # Changed the separator to a single space " "

# Create a StringIO object from the CSV string
fobj = StringIO(dataset_str)

# Use StringIO to simulate a file-like object
df = pd.read_csv(fobj, sep=" ", header=None, names=col_names)

# Separate features and target
X = df.iloc[:, :-1]  # All columns except the last one (features)
y = df.iloc[:, -1]   # The last column (target)

print("Features:")
print(X)
print("\nTarget:")
print(y)
```

Splitting the data into training and testing data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training : XGBoost Regressor

```python
#loading model
model = XGBRegressor()
```

```python
#training the model with X_train
model.fit(X_train,y_train)
```

Prediction on training data

```python
#accuracy for prediction on training data
training_data_pred = model.predict(X_train)
```

```python
print(training_data_pred)
```

```python
from sklearn import metrics
```

```python
# R squared error
score_1 = metrics.r2_score(y_train, training_data_pred)
print("R squared error : ", score_1)
```

```python
#Mean Absolute Error
score_2 = metrics.mean_absolute_error(y_train, training_data_pred)
print("Mean Absolute Error : ", score_2)
```

Visualisize Prediction and actual Prices

```python
plt.scatter(y_train, training_data_pred)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Prices vs Predicted Prices")
plt.show()
```

Prediction on test data

```python
test_data_pred = model.predict(X_test)
print(test_data_pred)
```

```python
#R squared Error
score_3 = metrics.r2_score(y_test, test_data_pred)
print("R squared error : ", score_3)

#Mean Absolute Error
score_4 = metrics.mean_absolute_error(y_test, test_data_pred)
print("Mean Absolute Error : ", score_4)
```

Visualize the test data

```python
plt.scatter(y_test, test_data_pred)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Prices vs Predicted Prices")
plt.show()
```
